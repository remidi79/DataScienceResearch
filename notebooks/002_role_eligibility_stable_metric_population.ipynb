# Experiment 002 — Role Eligibility & Stable Metric Population

## 1. Objective

Define reference populations for all future score engines: role eligibility, minutes thresholds, UNKNOWN/MULTI_ROLE handling, and metric stability screens.

## 2. Football hypothesis

Role-specific football scores are only defensible when populations are stable. Players with insufficient minutes or genuinely mixed roles should not define early model coefficients.

## 3. Dataset

DataPlatform source root: `/home/platform/DataPlatform/tmp/master_data_warehouse`. Uses direct StatsBomb player-season/player-match stats plus lineups for role-minute resolution.

````
import pandas as pd
summary = pd.read_csv('outputs/tables/002_role_eligibility_summary.csv')
summary
````

## 4. Feature engineering

Lineup position intervals are converted into role minutes. A player is assigned to a role only when one role has at least 70% of observed role minutes; otherwise the player becomes MULTI_ROLE. Eligibility thresholds are GK 600, CB/FB/MID 900, WINGER/CF 750 minutes.

````
pd.read_csv('outputs/tables/002_role_resolution.csv').head(20)
````

## 6. Statistical Analysis

For each role and metric: coverage, variance, coefficient of variation, bootstrap mean confidence interval, and split-half reliability from player-match data where possible.

````
stability = pd.read_csv('outputs/tables/002_metric_stability_by_role.csv')
stability.head(20)
````

## 7. Machine Learning

No coefficient model is trained in this experiment. The output defines the eligible modelling population and metric screen for later normalization, feature selection, and interpretable model experiments.

## 8. Explainability

Every metric has an explicit inclusion/exclusion reason: low coverage, near-zero variance, high correlation, high CV, or low split-half reliability.

## 9. Validation

The script validates required input files, writes deterministic tables/figures/reports, updates methodology append-only, and is checked by the ad-hoc verification command described in the final report.

## 10. Conclusions

Use eligible role populations and stable metric screens as gates before any score weighting. UNKNOWN and MULTI_ROLE players should not train initial coefficients.
