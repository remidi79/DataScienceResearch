# Experiment 003 — Scientific Feature Engineering & Normalization

## 1. Objective

Create the scientific feature layer for future score engines. No final weights and no supervised score model are built.

## 2. Football hypothesis

Role-specific ready metrics need role-specific normalization, benchmarks, redundancy checks, and latent-dimension discovery before score weighting.

## 3. Dataset

Uses Experiment 002 READY metrics only, player-season StatsBomb direct provider stats, and role eligibility from Experiment 002.

````
import pandas as pd
normalization = pd.read_csv('outputs/tables/003_normalization_decisions.csv')
normalization.head()
````

## 4. Feature engineering

For each eligible role, metrics are transformed by z-score, robust z-score, percentile rank, min-max, log transform, quantile transform, and winsorized z-score. The selected method minimizes transformed skew, excess kurtosis, and IQR outlier share.

````
pd.read_csv('outputs/tables/003_metric_statistics.csv').head(20)
````

## 5. Exploratory Data Analysis

Distribution, QQ, boxplot, and benchmark figures are produced per role under outputs/figures.

````
pd.read_csv('outputs/tables/003_role_benchmarks.csv').head(20)
````

## 6. Statistical Analysis

Pearson/Spearman correlations, redundancy pairs, hierarchical feature clustering, PCA variance, FactorAnalysis checks, and unsupervised feature-importance candidates are computed.

````
pd.read_csv('outputs/tables/003_feature_redundancy.csv').head(20)
````

````
pd.read_csv('outputs/tables/003_latent_dimensions.csv').head(20)
````

## 7. Machine Learning

No predictive model is trained. PCA, FactorAnalysis, QuantileTransformer, hierarchical clustering, and mutual information are used only for feature engineering research and weight preparation.

````
pd.read_csv('outputs/tables/003_weight_preparation.csv').head(20)
````

## 8. Explainability

Every normalization decision has an explicit distribution-quality score. Every latent dimension assignment includes cluster id and PCA/factor evidence.

## 9. Validation

The script validates required Experiment 002 inputs, regenerates all expected tables/figures/reports, and appends methodology once.

## 10. Conclusions

Experiment 003 prepares the normalized feature layer and benchmark definitions for Experiment 004. Final weights remain intentionally uncomputed.
